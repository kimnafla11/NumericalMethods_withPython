```python
import numpy as np
def f(x):
    return 0.95*(x**3) - (5.9)*(x**2) + 10.9*x - 6
def df(x):
    return 0.95 * (3 * x**2) - 5.9 * (2 * x) + 10.9
n = 8 #반복횟수
x = np.zeros(n) 
x[0] = 3.5 #초기값

for k in range(n-1):
    x[k+1] = x[k] -f(x[k])/df(x[k])

print("%5s %7s" %('k','x'))
for k in range(n):
    print("%5d %9.4f" %(k+1,x[k]))

```

```python
import numpy as np
def f(x):
    return 0.95*(x**3) - (5.9)*(x**2) + 10.9*x - 6

n = 10 #반복횟수
xs = np.zeros(n) #근 저장할 행렬
x0 = 2.5 #초기값
x1 = 3.5 #초기값

for k in range(n):
    x2 = x1 - f(x1)*((x1-x0)/f(x1)-f(x0))
    x0 = x1
    x1 = x2
    xs[k] = x2

print("%5s %8s" %('k','x'))
for k in range(n):
    print("%5d %9.4f" %(k+1,xs[k]))
```
