```python
#제 7장 : Matplotlib(시각화 라이브러리)

# 7.1. 함수를 그리는 법
# y = cos(x)를 그리는 프로그램
from numpy import linspace, cos,pi
import matplotlib.pyplot as plt
x = linspace(0,2.0*pi,40)
# 0에서 2*pi 사이의 40개 등간 배열
# -> 주어진 구간을 이산적인 데이터로 바꾸기 위해 linspace사용
y = cos(x) #코사인 함수에 대입
plt.plot(x,y,'k') #'k'는 플롯 색상인듯.
plt.xlabel('x')
plt.ylabel('y')
plt.title('y = cos(x)')
plt.show()
```

```python
# y = cos(x) 와 y = sin(x)를 그리는 프로그램
from numpy import linspace, cos, sin, pi
import matplotlib.pyplot as plt
x = linspace(0,2.0*pi,40)
# 0에서 2*pi 사이의 40개 등간 배열
# -> 주어진 구간을 이산적인 데이터로 바꾸기 위해 linspace사용
yc = cos(x) #코사인 함수에 대입
ys = sin(x) #사인 함수에 대입
plt.plot(x,yc,'ko-',label='cos(x)') #축에 대한 라벨링, 선 모양 표시 ko-는 검정색 점찍힌 선
plt.plot(x,ys,'k--',label='sin(x)') #k는 검정색 --는 점선
plt.legend() #범례를 표시함
plt.xlabel('x')
plt.ylabel('y')
plt.title('y = cos(x) & sin(x)')
plt.show()
```

```python
# 7.2 데이터를 사용하여 그리기
from numpy import array
import matplotlib.pyplot as plt
x = array([0,1,3,7,8])
y = array([1.1,2.5,-0.7,3.0,2.9])
plt.plot(x,y)
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

```python
# 7.4. 과제
# 1. y = sin(x)+cos(x)의 그래프를 0<=x<=2pi구간에서 그리시오
from numpy import pi, sin, cos
import matplotlib.pyplot as plt
x = linspace(0,2.0*pi,40)
y = sin(x)+cos(x)
plt.plot(x,y,'k')
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

```python
# 2. y = sin(2x)의 그래프를 0<=x<=2pi구간에서 그리시오
from numpy import sin, pi
import matplotlib.pyplot as plt
x = linspace(0,2.0*pi,40)
y = sin(2*x)
plt.plot(x, y, 'k')
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

```python
# 3. y=x/루트1+x**2의 그래프를 0<=x<=3구간에 그리기
import numpy as np
import matplotlib.pyplot as plt
x = np.linspace(0,3,50)
y = x/((1+x**2)**1/2)
plt.plot(x,y,'k')
plt.xlabel('x')
plt.ylabel('y')
plt.title('y = x/((1+x**2)**1/2)')
plt.show()
```

```python
# 과제4
x = np.linspace(0,5,40)
y = 3*(np.exp(-x))+x-1 #np.exp(0) e^0과 동일
plt.plot(x,y,'k')
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

```python
# 과제6
x = np.linspace(0,5,40)
y1 = 3*(np.exp(-x))+x-1
y2 = x/((1+x**2)**1/2)
plt.plot(x,y1,'k--', label = '3*(np.exp(-x))+x-1')
plt.plot(x,y2,'ko-', label = 'x/((1+x**2)**1/2)')
plt.legend()
plt.xlabel('x')
plt.ylabel('y')
plt.title('two plot')
plt.show()
```
